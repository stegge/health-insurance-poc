```python
#This step uses a for loop to read through the list of files, add the load timestamp and source columns before writing the data to a delta table.  
for df_name, file_path in files.items():
    df = spark.read.csv(file_path, inferSchema=True, header=True)
    df = df.withColumn("load_timestamp", current_timestamp()) \
           .withColumn("source", lit(file_path))
    df.write.format("delta").mode("append").saveAsTable(f"workspace.`tegge-insurance-data`.{df_name}")
```
